```python
#Problem 1: Mobile Recharge Validation
def recharge():
    plans = [199, 299, 399, 599]

    while True:
        amt = int(input("Enter amount: "))

        if amt < 50:
            print("Amount must be at least 50")
        elif amt not in plans:
            print("Invalid plan")
        else:
            print("Recharge successful")
            break

recharge()
```

```python
#Problem 2: Inventory Reorder Alert
def check_stock(stock):
    for item in stock:
        if stock[item] < 15:
            print(item, "Reorder Alert")
        else:
            print(item, "Stock OK")

products = {"Pen": 10, "Book": 20, "Eraser": 5}
check_stock(products)
```

```python
#Problem 3: Student Result Processing
def result(marks):
    total = 0
    for m in marks:
        total = total + m

    avg = total / len(marks)

    if avg >= 50:
        print("Pass")
    else:
        print("Fail")

marks = [40, 60, 50]
result(marks)
```

```python
#Problem 4: Cab Fare Estimator
def fare(km, peak):
    total = 50 + (km * 12)

    if peak == "yes":
        total = total + (total * 0.25)

    return total


while True:
    d = float(input("Enter km: "))
    p = input("Peak hour? yes/no: ")

    print("Fare =", fare(d, p))

    again = input("Again? yes/no: ")
    if again != "yes":
        break
```
