```python
#Problem 1: Post Engagement Analyzer
def analyze_likes(likes):
    total = 0
    for l in likes:
        total += l

    if total >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total)
    print("Post Status:", status)

analyze_likes([200, 300, 250, 300])
```

```python
#Problem 2: Medicine Stock Alert
def check_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)

check_stock(6)
```

```python
#Problem 3: Rainfall Adequacy Checker
def check_rainfall(data, required):
    total = 0
    for r in data:
        total += r

    avg = total / len(data)

    if avg >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", avg)
    print("Rainfall Status:", status)

check_rainfall([70, 75, 80, 63], 70)
```

```python
#Problem 4: Duplicate Account Detection
def check_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

check_duplicates(["ram", "shyam", "ram"])
```

```python
#Problem 5: Appointment Eligibility
def check_age(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)

check_age(21)
```

```python
#Problem 6: Premium Crop Price Filter
def premium_crops(prices):
    premium = []

    for p in prices:
        if p > 2000:
            premium.append(p)

    print("Premium Crops:", premium)

premium_crops([1500, 2500, 1800, 3200])
```

```python
#Problem 7: Application Health Checker
def check_errors(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

check_errors(7)
```

```python
#Problem 8: Transaction Limit Checker
def check_transaction(amount):
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

check_transaction(60000)
```

```python
#Problem 9: Attendance Eligibility
def check_attendance(days):
    total = len(days)
    present = 0

    for d in days:
        if d == 1:
            present += 1

    percentage = (present / total) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

check_attendance([1,1,1,0,1,1,1,1])
```

```python
#Problem 10: Electricity Bill Analyzer
def calculate_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Units Consumed:", units)
    print("Total Bill:", bill)
    print("Usage Status:", usage)

calculate_bill(250)
```
