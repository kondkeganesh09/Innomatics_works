```python
#Problem 1: Smart Parking Lot Management System
def parking_system(capacity, logs):
    count = 0
    for log in logs:
        if log == "IN":
            count += 1
        elif log == "OUT":
            count -= 1

    print("Currently Parked Vehicles:", count)

    if count >= capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")


# Sample Input
parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT","IN"])
```

```python
#Problem 2: Online Food Delivery Time Estimator
def delivery_time(distance, traffic, weather):
    time = distance * 5   # base time

    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10

    if weather == "Rainy":
        time += 5

    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
delivery_time(8, "High","MANSOON")
```

```python
#Problem 3: Movie Theatre Seat Occupancy Analyzer
def seat_analyzer(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")


# Sample Input
seat_analyzer(200, [1]*175)
```

```python
#Problem 4: Cloud Server Load Classification System
def server_load(cpu_readings):
    total = 0

    for cpu in cpu_readings:
        total += cpu

    average = total / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
server_load([45, 60, 70, 85, 90])
```

```python
#Problem 5: Smart Classroom Resource Usage Monitor
def resource_monitor(usage):
    overused = []

    for resource, hours in usage.items():
        if hours > 8:
            overused.append(resource)

    print("Overused Resources:", ", ".join(overused))

    if overused:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")


# Sample Input
resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```
