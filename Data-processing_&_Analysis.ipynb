```python
#Problem 1: Employee Performance Bonus Eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

max_score = max(employees.values())

for name in employees:
    if employees[name] == max_score:
        print(name, "(Score:", max_score, ")")
```

```python
#Problem 2: Search Query Keyword Analysis
text = "Buy mobile phone buy phone online"

text = text.lower()
words = text.split()

count = {}

for word in words:
    if word in count:
        count[word] = count[word] + 1
    else:
        count[word] = 1

for word in count:
    if count[word] > 1:
        print(word, ":", count[word])
```

```python
#Problem 3: Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid.append((i, sensor_readings[i]))

print(valid)
```

```python
#Problem 4: Email Domain Usage Analysis
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total = len(emails)

for domain in domain_count:
    percent = (domain_count[domain] / total) * 100
    print(domain, ":", int(percent), "%")
```
